```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('titanic.csv')
```

```python
# Probability of a person being a woman on the Titanic
float(len(df[df['Sex'] == 'female']))/len(df['Sex'].dropna())
```

```python
# Probability of a person on the titanic surviving   =>if<=   their Fair was greater than 100
float(len(df[(df['Survived'] == 1) & (df['Fare'] > 100)]))/len(df[df['Fare'] > 100])
```

```python
# probability of of surviving and their fair was more than 100
num = float(len(df[(df['Survived'] == 1) & (df['Fare'] > 100)]))/len(df[['Survived','Fare']].dropna())
num
```

```python
# The probability of a fare being greater than 100
den = float(len(df[df['Fare'] > 100]))/len(df[['Fare']].dropna())
den
```

```python
# verify Bayes rule: Probability of a person on the titanic surviving   =>if<=   their Fair was greater than 100
num/den
```

```python
# Given that you survived this is your chance of being male
float(len(df[(df['Sex'] == 'male') & (df['Survived'] == 1)]))/len(df[df['Survived'] == 1])
```

```python
# given that you are 20 < x < 30 whats the chance your Pclas is 1
float(len(df[(df['Age'] > 20) & (df['Age'] < 30) & (df['Pclass'] == 1)]))/len(df[(df['Age'] > 20) & (df['Age'] < 30)]['Pclass'].dropna())
```

```python
li = []
for year in df['founded_at'].dropna().sort_values().unique():
    li.append(df[df['founded_at'] == year]['funding_total_usd'].sum())
```

```python
sns.lineplot(y=li, x=df['founded_at'].dropna().sort_values().unique(), data=df)
```

```python
stwr = []
for year in df[df['category_list'] == 'Software']['founded_at'].dropna().sort_values().unique():
    stwr.append(df[(df['founded_at'] == year) & (df['category_list'] == 'Software')]['funding_total_usd'].sum())
```

```python
sns.lineplot(y=stwr, x=df[df['category_list'] == 'Software']['founded_at'].dropna().sort_values().unique(), data=df)
```
